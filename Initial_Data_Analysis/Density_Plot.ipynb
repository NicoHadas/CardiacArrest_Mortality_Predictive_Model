```python
# Density probability plot function
def density_plot(table, column):
    data = pd.read_csv(table)[column]
    if data.dtype != 'int64':
         data = pd.to_datetime(data)
    elif data.dtype != 'float64':
        data = pd.to_datetime(data) 
    plt.figure(figsize=(12, 6))
    sns.kdeplot(data, fill=False, color='red', alpha=0.75)
    sns.set_style("whitegrid")
    plt.title(f'Density Probability Plot for {column}')
    plt.xlabel(f'{column}')
    plt.ylabel('Density')
    return plt.show()
```

### admissions

```python
# Run density plot function# Univariate Distribution for the diagnosis_icd table columns
density_plot('admissions.csv', 'subject_id')
density_plot('admissions.csv', 'hadm_id')
density_plot('admissions.csv', 'dischtime')
density_plot('admissions.csv', 'deathtime')
density_plot('admissions.csv', 'edregtime')
density_plot('admissions.csv', 'edouttime')
```
